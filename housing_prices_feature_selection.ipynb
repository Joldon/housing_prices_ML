```python
# import dataset
data = pd.read_csv('data/iter-7/housing_prices.csv')
data
```

preprocess data

```python
# import train test split
from sklearn.model_selection import train_test_split
```

```python
# split data into train and test

X=data.drop(columns="Id")
y=X.pop("SalePrice")

# X = data.drop(['SalePrice'], axis=1)
# y = data['SalePrice']

# alternative way to split data into train and test
# X = data
# y = data.pop('SalePrice')

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
X
```

```python
#preprocessing pipeline
X_cat = X.select_dtypes(exclude="number").copy() #columns use here or later in preprocessor. 
X_num = X.select_dtypes(include="number").copy()# columns
```

create pipelines

```python


numeric_pipe = make_pipeline(SimpleImputer(strategy='mean'))

categoric_pipe = make_pipeline(SimpleImputer(strategy='constant', fill_value='N_A'), OneHotEncoder(handle_unknown='ignore', sparse_output=False))

preprocessor = ColumnTransformer(
    transformers=[
        ('num_pipe', numeric_pipe, X_num.columns),
        ('cat_pipe', categoric_pipe, X_cat.columns),
    ]
    )
```

```python
#this step is necessary in order to fit data into the model in the next step.
# without this step, the data won't preprocessed in fitting process


model_pipe_knr_minmax = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('minmax_scaler', MinMaxScaler()),
    # ('var_threshold', VarianceThreshold()),
    # ('mod_feature_sel', SelectFromModel(DecisionTreeRegressor())),
    ('model', KNeighborsRegressor())
])
```

```python
# Alternatively, minmax scaler make_pipeline

# model_pipe_knr_minmax = make_pipeline(preprocessor,
#                                       MinMaxScaler(),
#                                       KNeighborsRegressor())
```

```python
model_pipe_knr_minmax.fit(X_train, y_train)
```

```python
knn_pred_minmax = model_pipe_knr_minmax.predict(X_test)
```

```python
# import MaxAbsScaler


model_pipe_knr_maxabs = make_pipeline(
    preprocessor,
    MaxAbsScaler(),
    # ('var_threshold', VarianceThreshold()),
    # ('mod_feature_sel', SelectFromModel(DecisionTreeRegressor())),
    KNeighborsRegressor()
)
```

```python
model_pipe_knr_maxabs.fit(X_train, y_train)
```

```python
knn_pred_maxabs = model_pipe_knr_maxabs.predict(X_test)
```

```python
model_pipe_tree_minmax = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('my_scaler', MinMaxScaler()),
    # ('var_threshold', VarianceThreshold()),
    # ('mod_feature_sel', SelectFromModel(DecisionTreeRegressor())),
    ('model', DecisionTreeRegressor())
])
```

```python
model_pipe_tree_minmax.fit(X_train, y_train)
```

```python
tree_pred_minmax = model_pipe_tree_minmax.predict(X_test)
```

```python


model_pipe_tree_maxabs = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('my_scaler', MaxAbsScaler()),
    # ('var_threshold', VarianceThreshold()),
    # ('mod_feature_sel', SelectFromModel(DecisionTreeRegressor())),
    ('model', DecisionTreeRegressor())
])
```

```python
model_pipe_tree_maxabs.fit(X_train, y_train)
```

```python
tree_pred_maxabs = model_pipe_tree_maxabs.predict(X_test)
```
