A baseline dummy model

```python
from sklearn.model_selection import train_test_split

X = data[['GrLivArea']]
y = data['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#dummy model
def dummy_model(dimension):
    return dimension * 100
```

```python
# dummy model predictions
data['dummy_pred'] = dummy_model(data['GrLivArea'])
data
```

```python
# visualize dummy model
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GrLivArea', y='SalePrice')
sns.scatterplot(data=data, x='GrLivArea', y='dummy_pred')
plt.title('Relation between Price and GrLivArea')
plt.ylabel('Price($)')
sns.despine()
```

```python
# calculate dummy model error with Root Mean Squared Error(RMSE)
from sklearn.metrics import mean_squared_error

y_train_pred = dummy_model(X_train['GrLivArea'])
y_test_pred = dummy_model(X_test['GrLivArea'])

train_rmse = mean_squared_error(y_train, y_train_pred, squared=False)
test_rmse = mean_squared_error(y_test, y_test_pred, squared=False)

print(f'Train RMSE: {train_rmse: .2f}')
print(f'Test RMSE: {test_rmse: .2f}')
```

Visualization of error/residuals

```python
import matplotlib.pyplot as plt

def plot_errors(X, y_true, y_pred, title, figsize=(10, 6), alpha=0.5, color='red', sample_size=50):
    #sample data for plotting purposes
    X_train_sample = X.sample(n=sample_size, random_state=42)
    y_train_sample = y_true.loc[X_train_sample.index]
    y_train_pred_sample = y_pred.loc[X_train_sample.index]

    #Plot data
    fig, ax = plt.subplots(figsize=figsize)
    ax.scatter(X_train_sample, y_train_sample, alpha=alpha, label='Actual Prices')
    ax.scatter(X_train_sample, y_train_pred_sample, alpha=alpha, label='Predicted Prices', color=color)

    #draw error lines
    for x, y_t, y_p in zip(X_train_sample.values.flatten(), y_train_sample, y_train_pred_sample):
        ax.plot([x, x], [y_t, y_p], color='gray', linestyle='--', linewidth=0.5)

    #plot details
    title = title + f' (Sample size: {sample_size})'
    ax.set_title(title)
    ax.set_xlabel('GrLivArea')
    ax.set_ylabel('SalePrice')
    ax.legend()
    plt.show()
```

Linear Regression

```python
#Linear regression is based on the equation:
# price = a * carat + b
# a = slope, b = intercept

from sklearn.linear_model import LinearRegression
 #create and fit model
lr = LinearRegression()
lr.fit(X_train, y_train)

#Make predictions
y_train_pred_lr = lr.predict(X_train)
y_test_pred_lr = lr.predict(X_test)
```

```python
# Evaluate linear regression model using the RMSE metric
train_rmse_lr = mean_squared_error(y_train, y_train_pred_lr, squared=False)
test_rmse_lr = mean_squared_error(y_test, y_test_pred_lr, squared=False)

print(f'Linear Regression Train RMSE: {train_rmse_lr: .2f}')
print(f'Linear Regression Test RMSE: {test_rmse_lr: .2f}')
```

```python
#compare dummy model and linear regression model
print('RMSE Evolution:')
print('-' * 30)
print(f'1. Dummy Model: \n\tTrain RMSE: {train_rmse: .2f}\n\tTest RMSE: {test_rmse: .2f}')
print('-' * 30)
print(f'2. Linear Regression Model (GrLivArea): \n\tTrain RMSE: {train_rmse_lr: .2f}\n\tTest RMSE: {test_rmse_lr: .2f}')
```

Multivariate regression

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
```

```python
# preprocessing pipeline

X_all = data
y_all = data.pop('SalePrice')
```

```python
X_train_all, X_test_all, y_train_all, y_test_all = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
```

```python
X_cat = X_all.select_dtypes(exclude="number").copy()
X_num = X_all.select_dtypes(include="number").copy()
```

```python
numeric_pipe = make_pipeline(
    SimpleImputer(strategy="mean"))
 
categoric_pipe = make_pipeline(
    SimpleImputer(strategy="constant", fill_value="N_A"),
    OneHotEncoder(handle_unknown="ignore")
)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num_pipe", numeric_pipe, X_num.columns),
        ("cat_pipe", categoric_pipe, X_cat.columns),
    ]
)
```

```python
reg_pipe = make_pipeline(preprocessor,
                         StandardScaler(with_mean=False, with_std=True),
                         LinearRegression()
                         )
```

```python
reg_pipe.fit(X_train_all, y_train_all)
```

```python
#Make predictions
y_train_pred_all = reg_pipe.predict(X_train_all)
y_test_pred_all = reg_pipe.predict(X_test_all)
```

```python
# Evaluate the model using R2, RMSE and MAE
train_rmse_all = mean_squared_error(y_train_all, y_train_pred_all, squared=False)
test_rmse_all = mean_squared_error(y_test_all, y_test_pred_all, squared=False)

print(f'Improved Model Train RMSE: {train_rmse_all:.2f}')
print(f'Improved Model Test RMSE: {test_rmse_all:.2f}')
```

```python
print('RMSE Evolution')
print('-' * 30)
print(f'1. Dummy Model: \n\tTrain RMSE: {train_rmse:.2f}\n\tTest RMSE: {test_rmse:.2f}')
print('-' * 30)
print(f'2. Linear Regression Model (GrLivArea): \n\tTrain RMSE: {train_rmse_lr:.2f}\n\tTest RMSE: {test_rmse_lr:.2f}')
print('-' * 30)
print(f'3. Improved Model (All features): \n\tTrain RMSE: {train_rmse_all:.2f}\n\tTest RMSE: {test_rmse_all:.2f}')
```

```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

train_mape_all = mean_absolute_percentage_error(y_train_all, y_train_pred_all)
test_mape_all = mean_absolute_percentage_error(y_test_all, y_test_pred_all)

print(f"Improved Model Train MAPE: {train_mape_all:.2f}%")
print(f"Improved Model Test MAPE: {test_mape_all:.2f}%")
```
